```python
#Showing Multiple relationship 
#Grouping demensions 
# col > splitting in horizontal direction
#row > splitting in vertical direction

import seaborn as sns

tips = sns.load_dataset('tips')

```

```python
sns.relplot( x = 'total_bill', y='tip', hue = 'sex', size = 'day', style = 'time', data=tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'sex', col = 'day', row ='time', data=tips)
```

# Ploting Categorical Varibles

```python
#relplot() > relationship between numerical vairable
#catplot() > categorical plotting 
```

```python
sns.catplot( x = 'day', y= 'tip',hue = 'sex',  data=tips)
```

```python
sns.relplot( x = 'day', y= 'tip', data=tips)
```

# order


```python
sns.catplot(x = 'day', y='tip', hue = 'sex', order = ['Sun','Sat', 'Thur', 'Fri'], data = tips)
```

# Visulizing the distribution 

#box and boxen 
#violin

Box-plot 
- descriptive statistics 

```python
sns.catplot(x="day", y="total_bill", order = ['Sat','Sun','Thur', 'Fri'], kind = 'box', hue = 'sex', data = tips)
```

# Descriptive Statistics for individual variables

```python
sns.displot(tips['tip']) #for numeric variable
```

# Histogram

```python
#Joining histogram and scatter plots 
sns.jointplot(x='total_bill', y="tip", data=tips)
```

# Visualizing pairwise correlations 

```python
sns.pairplot(tips)
```
