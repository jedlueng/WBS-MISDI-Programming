## Simple Line Plotting

```python
#creating some sample datasets
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(0, 100, 50) #Return evenly spaced numbers over a specified interval. (start,stop,num)
y1 = 2 * x
y2 = 3 * x
y3 = 2 * x + np.random.normal(0, 10, 50)
```

### Plot the lines with pyplot API

```python
#plot two datasets in one figure
plt.figure()
#Plot Line one
plt.plot(x, y1)
#Plot Line two 
plt.plot(x, y2)
#uncomment the line below when working with IDEs
#plot.show()
```

```python
#plot two datasets in two figures
#pyplot interface executes commands sequentially
#Object Orenited 
plt.figure() #Figure 1 
plt.plot(x, y1)#Plot line 1 on figure 1
plt.figure()#Figure 2 
plt.plot(x, y2)#Plot line 2 on figure 2 
#uncomment the line below when working with IDEs
#plot.show()
```

### Plot the lines with Object-oriented API

```python
#plot two datasets in one figure
fig1 = plt.figure()
ax1 = plt.axes()
ax2 = plt.axes()
ax1.plot(x, y1)
ax2.plot(x, y2)
#uncomment the line below when working with IDEs
#plot.show()
```

```python
#plot two datasets in two figures
#Figure 1 
fig1 = plt.figure()
ax1 = plt.axes()
ax1.plot(x, y1)
#Figure 2
fig2 = plt.figure()
ax2 = plt.axes()
ax2.plot(x, y2)
#uncomment the line below when working with IDEs
#plot.show()
```

## Customize your figures

### Anatomy of a figure
anatomy.png

### Set figure size

```python
#set figure size
plt.figure(figsize=(3, 4))
plt.plot()
```

### Set axis limits

```python
#Sex limit on axis 
plt.xlim((-50, 100))  #tuple as one argument
plt.ylim((-50, 300))
```

### Set axis label

```python
#Setting axis label
plt.xlabel('This is x axis')
plt.ylabel('This is y axis')
```

### Set axis ticks

```python
#Customized tick 
plt.xlim((-50, 100))
plt.ylim((-50, 300))
plt.xticks([-50, 0, 50, 100])
plt.yticks([-50, 50, 150, 300])
```

```python
#Customaize tick with letters 
plt.xlim((-50, 100))
plt.ylim((-50, 300))
plt.xticks([-50, 0, 50, 100], ['A', 'B', 'C', 'D'])
plt.yticks([-50, 50, 150, 300])
```

### Add legend

```python
#Add a legend for each line 
plt.figure()
plt.plot(x, y1, label='y=2x')
plt.plot(x, y2, label='y=3x')
plt.legend(loc='lower right')
```

```python
plt.figure()
'''
return a list of lines representing the plotted data.
make sure you have the comma
'''
line1, = plt.plot(x, y1)
line2, = plt.plot(x, y2)
#Auto generating 
plt.legend(handles=[line1, line2], labels=['y=2x', 'y=3x'], loc='best')
```

### Add text


```python
plt.figure()
plt.plot(x, y1)
plt.text(20, 150, 'Some explanation') #All text to the figure
```

## Add annotation

```python
#Same as adding text 
plt.figure()
plt.plot(x, y1)
x0 = 40
y0 = 2 * x0
plt.annotate(
    'test',
    xy=(x0, y0),
    xycoords='data',
    xytext=(+30, -30),
    textcoords='offset points',
    fontsize=16,
)
```

## Move axis

```python
plt.figure()
plt.plot(x, y1)
ax = plt.gca()  #get current axis
#Setting color and positions 
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))  #a tuple as one argument
ax.spines['left'].set_position(('data', 0))
```

## Save your figure into file

```python
plt.figure()
plt.plot(x, y1)
ax = plt.gca()  #get current axis
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))  #a tuple as one argument
ax.spines['left'].set_position(('data', 0))
plt.xlim((-50, 50))  #tuple as one argument
plt.ylim((-100, 100))
plt.xlabel('xxx')
plt.ylabel('yyy')
plt.xticks([-50, -25, 0, 25, 50])
plt.yticks([-100, -50, 0, 50, 100])
plt.legend(labels=['y=2x'], loc='best')
x0 = 40
y0 = 2 * x0
plt.annotate('x = 40\ny = 80',
             xy=(x0, y0),
             xycoords='data',
             xytext=(+30, -30),
             textcoords='offset points',
             fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=.2"))
plt.text(15,
         -50,
         'A simple line plotting demo',
         fontdict={
             'size': 16,
             'color': 'r'
         })
plt.savefig("figure1.png", transparent=True, bbox_inches='tight') #Saving figure here 
```

## Scatter Plotting

```python
plt.scatter(x, y3, s=100, c='#000000', marker='>', alpha=0.5)
#https://matplotlib.org/3.3.3/api/markers_api.html#module-matplotlib.markers for different marker styles
```

## Bar Plotting

```python
plt.bar(x, y1, facecolor='#000000', edgecolor='white')
for m, n in zip(x, y1):
    # ha: horizontal alignment
    # va: vertical alignment
    plt.text(m + 0.4, n + 0.05, '%.2f' % n, ha='center', va='bottom')
```

# Customize Seaborn plot

```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

## Set plot before Seaborn visualization

```python
plt.figure(figsize=(4, 8))
plt.ylim((0, 200))
sns.histplot(x='day', data = tips)
```

## Update Seaborn plot after visualization

```python
sns.relplot(x='total_bill', y='tip', data = tips)
plt.gcf().set_size_inches(12, 8)
```

## Similar customization with Seaborn parameters

```python
sns.relplot(x='total_bill', y='tip', data = tips, height = 8, aspect = 12/8)
```

```python
plt.figure(figsize=(4, 8))
sns.regplot(x='total_bill', y='tip', data = tips)#regplot is an axes function
```

```python
sns.relplot(x='total_bill', y='tip', data = tips)#relplot is a figure function
plt.ylim((0, 100))
```
