# Exercise 1

## Generate x and y

```python
import numpy as np
x = np.linspace(-10,10,100)
y1 = 1 - 2*x + x**2
```

## plotting with pyplot

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(x,y1)
```

## Object-oriented plotting

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x,y1)
```

```python
#you can create the figure object and axes object at the same time 
#by calling pyplot.subplots()
fig, ax = plt.subplots()
ax.plot(x,y1)
```

## Exercise 1-2

```python
y2 = 26 - 2 * x
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.figure(figsize=(10,5))
plt.xlim(-15,15)
plt.xticks(range(-15,20,5))
plt.xlabel('x from -15 to 15')
plt.ylabel('y from 0 to 120')
plt.title('Line Intersection Plot')
plt.text(-5,80,'There are two intersections between two lines.')
plt.plot(x,y1,label='y=x**2-2x+1')
plt.plot(x,y2,label='y=26-2*x')
plt.legend(loc='lower left')
```

## Exercise 1-3

```python
plt.figure(figsize=(10,5))
plt.xlim(-16,16)
plt.xticks(range(-15,20,5))
plt.xlabel('x from -15 to 15')
plt.ylabel('y from 0 to 120')
plt.title('Line Intersection Plot')
plt.text(2,90,'There are two intersections between two lines.')
plt.plot(x,y1,label='y=x**2-2x+1')
plt.plot(x,y2,label='y=26-2*x')
plt.legend()
plt.annotate('(-5,36)', xy=(-5,36), xytext=(-30, -30),textcoords='offset points',
            arrowprops=dict(arrowstyle="->", color='black'))
plt.annotate('(5,16)', xy=(5,16), xytext=(-30, +30),textcoords='offset points',
            arrowprops=dict(arrowstyle="->", color='black'))
ax = plt.gca()  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
```
