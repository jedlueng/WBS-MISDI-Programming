Simple plot for x and y1 

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10,10,50)
y1 = x**2 - 2*x +1
y2 = 26 -2 * x
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x, y1, label='y1 = x**2 â€“ 2*x + 1')
plt.plot(x, y2, label='y2 = 26 -2 * x')
plt.legend(loc='lower left')
plt.text(-6,90, 'There\'re two intersection between two lines')
plt.title('Line intersection plot')
plt.xlim(-15,15)
plt.xlabel('x from -15 to 15') #message in the paremeter 
plt.ylabel('y form 0 to 120')
plt.style.use('seaborn')
```

# Object oriented way 

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x,y1)
ax.plot(x,y2)
plt.style.use('seaborn')
```
