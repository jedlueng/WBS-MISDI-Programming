```python
import pandas as pd
import seaborn as sns 
games = pd.read_csv('all_games.csv')
```

```python
#show dataframe
games.head()
```

```python
df1 = games[['platform', 'meta_score']]
```

```python
#df2 is a average group by platform 
df2 = df1.groupby('platform').mean()
df2.head()
```

```python
sns.scatterplot(data=df2, x="platform", y="meta_score").set_title('Average Metascore recieve by Companies')
```

```python
sns.relplot(x='year', y='user_review', kind = 'line', hue = 'platform', data=games)
```
