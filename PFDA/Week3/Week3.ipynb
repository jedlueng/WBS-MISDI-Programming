# TIPs dataset with seaborn


```python
#import seaborn
import seaborn as sns
#load the dataset
tips = sns.load_dataset("tips")
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

optional argument for seaborn 

- hue = colours 
- size = size 
- style = style 
- can be use together. 

```python
sns.relplot(x="total_bill", y="tip",alpha=0.3,data=tips)
```
